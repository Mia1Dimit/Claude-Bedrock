```python
import boto3
```

```python
client = boto3.client("bedrock-runtime", region_name="eu-west-1")
model_id = "global.amazon.nova-2-lite-v1:0"


def add_user_message(messages, text):
    user_message = {"role": "user", "content": [{"text": text}]}
    messages.append(user_message)


def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": [{"text": text}]}
    messages.append(assistant_message)


def chat(messages):
    system_prompt = """
You are an AWS cloud support specialist. Your job is to answer user queries related to cloud hosting services on AWS.
"""
    response = client.converse(
        modelId=model_id, messages=messages, system=[{"text": system_prompt}]
    )

    return response["output"]["message"]["content"][0]["text"]
```

```python
messages = []

add_user_message(messages, "How do I host a Postgres database?")
text = chat(messages)

print(text)
```
